```python
import pandas as pd
```

```python
file = 'Resources/first_round_csvs/AllFirstRoundDrafts.csv'
all_players = pd.read_csv(file)
all_players
```

```python
all_players.columns
```

```python
print(all_players['Pos'])
```

```python
all_players.groupby(["Pos"]).count()['Year']
```

```python
new_col_order = ['Name','Year', 'OvPck', 'RdPck', 'Rnd','Tm', 'Pos', 'DT', 'FrRnd','Signed', 'Bonus',
                 'WAR', 'G', 'AB', 'HR', 'BA', 'OPS', 'G.1', 'W', 'L',
       'ERA', 'WHIP', 'SV', 'Type', 'Drafted Out of']
```

```python
all_players = all_players[new_col_order]
all_players
```

```python
pitchers = all_players.loc[(all_players['Pos'] == 'LHP') | (all_players['Pos'] == 'RHP')]
pitchers
```

```python
batters = all_players.loc[(all_players['Pos'] != 'LHP') & (all_players['Pos'] != 'RHP')]
batters
```

```python
batters = batters.drop(['G.1', 'W', 'L',
       'ERA', 'WHIP', 'SV', 'DT'], 1)
```

```python
batters = batters.fillna(0)
batters
```

```python
batters.to_csv('Resources/first_round_csvs/FirstRoundBatters.csv', index=False)
```

```python
pitchers = pitchers.drop(['G', 'AB', 'HR', 'BA', 'OPS', 'DT'], 1)
pitchers = pitchers.fillna(0)
pitchers
```

```python
pitchers.to_csv('Resources/first_round_csvs/FirstRoundPitchers.csv', index=False)
```

```python
all_players = all_players.sort_values(by=['Year']).reset_index()
all_players
```

```python
all_players.to_csv('Resources/first_round_csvs/AllFirstRoundDrafts.csv', index=False)
```
