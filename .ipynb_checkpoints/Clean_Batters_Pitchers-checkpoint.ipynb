```python
import pandas as pd
import numpy as np
import glob
import os
import re
```

```python
file_path = 'Resources/AllFirstRoundDrafts_Clean.csv'
player_df = pd.read_csv(file_path)
player_df
```

```python
player_df.columns
```

```python
print(player_df['Pos'])
```

```python
player_df.groupby(["Pos"]).count()['Year']
```

```python
pitchers = player_df.loc[(player_df['Pos'] == 'LHP') | (player_df['Pos'] == 'RHP')]
pitchers
```

```python
batters = player_df.loc[(player_df['Pos'] != 'LHP') & (player_df['Pos'] != 'RHP')]
batters
```

```python
batters = batters.drop(['G.1', 'W', 'L',
       'ERA', 'WHIP', 'SV'], 1)
batters
```

```python
batters.to_csv('Resources/cleanBatters.csv', index=False)
```

```python
pitchers = pitchers.drop(['G', 'AB', 'HR', 'BA', 'OPS'], 1)
pitchers = pitchers.rename(columns = {'G.1':'G'})
pitchers
```

```python
pitchers.to_csv('Resources/cleanPitchers.csv', index=False)
```

```python
no_mlb_pitchers = pitchers.loc[(pitchers['WAR'] == 0) & (pitchers['G.1'] == 0) & (pitchers['W'] == 0) &(pitchers['L'] == 0)
                     & (pitchers['ERA'] == 0) & (pitchers['WHIP'] == 0) & (pitchers['SV'] == 0)]
no_mlb_pitchers
```

```python
no_mlb_batters = batters.loc[(batters['WAR'] == 0) & (batters['G'] == 0) & (batters['AB'] == 0) &(batters['HR'] == 0)
                     & (batters['BA'] == 0) & (batters['OPS'] == 0)]
no_mlb_batters 
```
