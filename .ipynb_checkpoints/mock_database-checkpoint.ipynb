```python
import pandas as pd
```

```python
file_to_load = "Resources/mock_data2.csv"
mock_df = pd.read_csv(file_to_load)
mock_df.head()
```

```python
field_players_df = mock_df.drop(['Team_Make_Playoffs_Year1', 'CareerLengthSeasons'], axis=1)
field_players_df.head()
```

```python
col = ['PlayerName', 'Team', 'Pos', 'Year', 'First_Round_Pick_Num', 'G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI']
field_players_df = field_players_df[col]
field_players_df.head()
```

```python
from sqlalchemy import create_engine
from config import db_password

db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/baseball_draft"
engine = create_engine(db_string)
```

```python
field_players_df.to_sql(name='field_players', con=engine, if_exists='replace')
```

```python
longevity_df = mock_df.filter(['PlayerName', 'Pos', 'CareerLengthSeasons'])
longevity_df.head()
```

```python
longevity_df.to_sql(name='longevity', con=engine, if_exists='replace')
```
